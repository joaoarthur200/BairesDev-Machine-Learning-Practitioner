```python
import cv2
from matplotlib import pyplot as plt
from mtcnn import MTCNN

# Caminho da imagem
image_path = '/content/png-clipart-organization-rotary-international-experience-business-psychology-group-of-people-miscellaneous-service.png'

# Carregar a imagem
image = cv2.imread(image_path)

# Verificar se a imagem foi carregada corretamente
if image is None:
    print("Erro ao carregar a imagem. Verifique o caminho ou o formato do arquivo.")
else:
    # Converter a imagem de BGR (formato do OpenCV) para RGB
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Inicializar o detector de faces MTCNN
    detector = MTCNN()

    # Detectar as faces na imagem
    faces = detector.detect_faces(image_rgb)

    # Filtrar as deteces: somente considerar faces com uma rea mnima
    min_face_area = 500  # rea mnima da face, pode ajustar conforme necessrio
    valid_faces = [face for face in faces if face['box'][2] * face['box'][3] > min_face_area]

    # Exibir a imagem com as faces detectadas
    plt.imshow(image_rgb)
    ax = plt.gca()

    # Desenhar as caixas delimitadoras das faces vlidas
    for face in valid_faces:
        x, y, width, height = face['box']
        ax.add_patch(plt.Rectangle((x, y), width, height, fill=False, color='red', linewidth=2))

    plt.axis('off')
    plt.show()
```
